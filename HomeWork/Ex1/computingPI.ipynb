##  Вычисление числа pi с помощью метода Монте-Карло

### Формулы
- Площадь квадрата: $s = (2 r)^2$
- Площадь круга: $c = \pi r^2$
- $c/s = (\pi r^2) / (4 r^2) = \pi / 4$
- $\pi = 4 * c/s$

### Импортируем модули

```python
import random
import matplotlib.pyplot as plt
#print("helloll")
```

### Инициализируем количесвто точек

```python
num_points = 5000
```

### Броски дротиков

```python
points = []
hits = 0
for _ in range (num_points):
    x, y = random.random(), random.random()
    if x*x + y*y < 1.0:
        hits += 1
        points.append((x,y, "green"))
    else:
        points.append((x,y, "orange"))
```

### Построение графика

```python
x, y, colors = zip(*points)
fig, ax = plt.subplots()
fig.set_size_inches(6.0,6.0)
ax.scatter(x,y, c = colors)

fraction = hits / num_points
print('Число pi = ' + str(4 * fraction))
```

## Хрень2

```python
from matplotlib import pyplot
import random
from ipywidgets import interact, widgets
```

```python
%matplotlib inline


def throw_darts(num_points):
    points = []
    hits = 0
    for _ in range(num_points):
        x, y = random.random(), random.random()
        if x*x + y*y < 1.0:
            hits += 1
            points.append((x, y, "green"))
        else:
            points.append((x, y, "orange"))
    fraction = hits / num_points
    pi = 4 * fraction
    return pi, points


def create_plot(points):
    x, y, colors = zip(*points)
    pyplot.scatter(x, y, c=colors)


def experiment(num_points):
    pi, points = throw_darts(num_points)
    create_plot(points)
    print("approximation:", pi)
```

```python
interact(experiment, num_points=widgets.IntSlider(min=100, max=10000, step=100,
value=4800))
```

### Подгоняем данные
Полином 5-го порядка к гауссовой
функции с некоторым случайным шумом

```python
import numpy as np

import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
%matplotlib inline


def gaussian(x, a, b, c):
    return a * np.exp(-b * (x-c)**2)


def noisy_gaussian():
    # gaussian array y in interval -5 <= x <= 5
    nx = 100
    x = np.linspace(-5.0, 5.0, nx)
    y = gaussian(x, a=2.0, b=0.5, c=1.5)
    noise = np.random.normal(0.0, 0.2, nx)
    y += noise
    return x, y


def fit(x, y, n):
    pfit = np.polyfit(x, y, n)
    yfit = np.polyval(pfit, x)
    return yfit


def plot(x, y, yfit):
    plt.plot(x, y, "r", label="Data")
    plt.plot(x, yfit, "b", label="Fit")
    plt.legend()
    plt.ylim(-0.5, 2.5)
    plt.show()


x, y = noisy_gaussian()
yfit = fit(x, y, n=5)  # fit a 5th order polynomial to it
plot(x, y, yfit)
```
